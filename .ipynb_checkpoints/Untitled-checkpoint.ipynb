```python
# imported libraries
import cv2 as cv
import matplotlib.pyplot as plt
from convert import dec_to_bin, bin_to_dec
from copy import copy as cpy
```

```python
# read an image file
img = cv.imread('testimg/4.1.04.tiff', cv.IMREAD_GRAYSCALE)
```

```python
# showing image using matplotlib
plt.imshow(img, cmap='gray')
plt.title('original')
```

```python
# getting image height and width
height = img.shape[0]
width = img.shape[1]
print("height: {}, width: {}".format(height, width))
```

```python
# obtaining bitmap8
bitplane8 = cpy(img)
bitplane7 = cpy(img)
bitplane6 = cpy(img)
bitplane5 = cpy(img)
bitplane4 = cpy(img)
bitplane3 = cpy(img)
bitplane2 = cpy(img)
bitplane1 = cpy(img)
bitplanes = [bitplane8, bitplane7, bitplane6, bitplane5, bitplane4, bitplane3, bitplane2, bitplane1]

for a in range(len(bitplanes)):
    for i in range(height):
        for j in range(width):
            bin_val = dec_to_bin(img[i,j])
            if str(bin_val)[2] == '1':
                bitplanes[a][i,j] = 255
            else:
                bitplanes[a][i,j] = 0
```

```python
titles = ['original', 'bitmap8']
images = [img, bitmap8]
# titles = ['original', 'bitplane8', 'bitplane7', 'bitplane6', 'bitplane5', 'bitplane4', 'bitplane3', 'bitplane2', 'bitplane1']
# images = [img, bitplane8, bitplane7, bitplane6, bitplane5, bitplane4, bitplane3, bitplane2, bitplane1]
```

```python
# viewing bitplane images
for i in range(len(images)):
    plt.subplot(1,2,i+1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap='gray', vmin=0, vmax=255)
    plt.xticks([])
    plt.yticks([])
```
