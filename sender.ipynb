```python
# imported libraries
import cv2 as cv
import matplotlib.pyplot as plt
from calculation import *
```

```python
# read an image file
img = cv.imread('test-images/miscellaneous/4.1.04.tiff', cv.IMREAD_GRAYSCALE)
# img = cv.imread('test-images/aerials/2.1.01.tiff', cv.IMREAD_GRAYSCALE)
```

```python
# showing image using matplotlib
plt.imshow(img, cmap='gray')
plt.title('original')
```

```python
# getting image height and width
height = img.shape[0]
width = img.shape[1]
print("height: {}, width: {}".format(height, width))
```

## Bitplanes decomposition

```python
# obtaining bitplanes
bp8, bp7, bp6, bp5, bp4, bp3, bp2, bp1 = bitplane_decomposition(img, height, width)
```

```python
titles = ['original', 'bitplane8', 'bitplane7', 'bitplane6', 'bitplane5', 'bitplane4', 'bitplane3', 'bitplane2', 'bitplane1']
images = [img, bp8, bp7, bp6, bp5, bp4, bp3, bp2, bp1]
```

```python
# viewing bitplane images
for i in range(len(images)):
    # convert inner pixels 1 to 255
    images[i][images[i]==1] = 255
    # displaying the images in subplot
    plt.subplot(3,3,i+1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap='gray', vmin=0, vmax=255)
    plt.subplots_adjust(bottom=0.1, right=1.5, top=2.2)
```

## Generating cipher images

```python
# getting cypher images for bitplane images up to its significance level
x = 0.78
y = 0.98
u = 3.57
cipher_images = []
for i in range(4):
    x, y, random_matrix = get_random_matrix(u, x, y, height, width)
    cipher_image = xor_matrix(random_matrix, images[i+1])
    cipher_images.append(cipher_image)
```

```python
# displaying all cyper images
plt.subplot(1,2,1)
plt.title('bitplane8')
plt.imshow(images[1], cmap='gray', vmin=0, vmax=255)
plt.subplot(1,2,2)
plt.title('cipher image')
plt.imshow(cipher_images[0], cmap='gray', vmin=0, vmax=255)
plt.subplots_adjust(bottom=0.1, right=1.5, top=2.2)
```
